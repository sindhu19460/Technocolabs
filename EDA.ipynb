```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:\\Users\\User\\Downloads\\fifa.csv")
```

```python
data.head()
```

```python
data.isnull().any()
```

```python
plt.figure(figsize=(16,6))

sns.lineplot(data=data)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(data=data)
plt.show()
```

```python
sns.lmplot(x='BRA', y='GER', data=data,
           fit_reg=False, 
           hue='ITA')
 
# Tweak using Matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
sns.boxplot(data=data)
```

```python
sns.set_style('whitegrid')
 #darkgrid,dark,white,ticks
# Violin plot
sns.violinplot(x='BRA', y='ITA', data=data)
plt.show()
```

```python
corr = data.corr()
sns.heatmap(corr)
```

```python
sns.displot(data)
```

```python
sns.swarmplot(x='BRA', y='ITA', data=data)
```

```python
sns.lmplot(x='BRA', y='ITA', data=data)
plt.show()
```

```python
sns.distplot(a = data['BRA'],kde=False)
```

```python
sns.kdeplot(data=data['BRA'],shade=True)
```

```python
# 2D KDE plot
sns.jointplot(x=data['BRA'], y=data['ITA'], kind="kde",shade=True)
plt.show()
```
