```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor as rf
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.describe()
```

```python
# Obtain target and predictors
y = train.SalePrice
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = train[features].copy()
X_test = test[features].copy()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
x_train.head()
```

```python
model_1 = rf(n_estimators = 50, random_state=0)
model_2 = rf(n_estimators=100, random_state=0)
model_3 = rf(n_estimators=100, criterion='mae', random_state=0)
model_4 = rf(n_estimators=200, min_samples_split=20, random_state=0)
model_5 = rf(n_estimators=100, max_depth=7, random_state=0)

models = [model_1, model_2, model_3, model_4, model_5]
```

```python
from sklearn.metrics import mean_absolute_error 
def score_model(model, X_t=x_train, X_v=x_test, y_t=y_train, y_v=y_test):
    model.fit(X_t, y_t)
    preds = model.predict(X_v)
    return mean_absolute_error(y_v, preds)

for i in range(0, len(models)):
    mae = score_model(models[i])
    print("Model %d MAE: %d" % (i+1, mae))
```

```python
best_model = model_3
```

```python
my_model = best_model
```

```python
my_model.fit(X,y)
```

```python
pred = my_model.predict(x_test)
print(pred)
```

```python
output = pd.DataFrame({'Id':x_test.index,
                      'SalePrice ':pred})

output.to_csv("submission.csv",index = False)
```
